```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
shop = pd.read_csv("/content/online_shoppers_intention.csv")
shop.head()
```

```python
shop.isnull().sum()
```

```python
shope = shop.drop(['Administrative', 'Informational', 'ProductRelated_Duration'], axis = 1)
```

```python
visitor = pd.get_dummies(shope['VisitorType'])

X = pd.concat([shope, visitor], axis=1)
X.head()
```

```python
y = X['Revenue']
X_new = X.drop(['Revenue', 'Month', 'VisitorType'], axis = 1)
X_new
```

```python
X_new['Weekend'] = np.asarray(X_new['Weekend']).astype(np.float32)

y = np.asarray(y).astype(np.float32)
```

```python
model = Sequential()
model.add(Dense(units=18,activation="relu"))
model.add(Dense(units=32,activation="relu"))
model.add(Dense(units=16,activation="relu"))
model.add(Dense(units=2,activation="softmax"))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_new,y,epochs=100, validation_split=0.1)
```

```python


y = X['Revenue']
X_new = X.drop(['Revenue', 'Month', 'VisitorType'], axis=1)

X_new['Weekend'] = X_new['Weekend'].astype(np.float32)


X_new = X_new.astype(np.float32)

y = y.astype(np.float32)

# Convert DataFrame/Series to NumPy array
X_new = X_new.values
y = y.values

# Check for NaN values
assert not np.isnan(X_new).any(), "X_new contains NaN values"
assert not np.isnan(y).any(), "y contains NaN values"


model = Sequential()
model.add(Dense(units=18, activation="relu", input_shape=(X_new.shape[1],)))
model.add(Dense(units=32, activation="relu"))
model.add(Dense(units=16, activation="relu"))
model.add(Dense(units=2, activation="softmax"))


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(X_new, y, epochs=100, validation_split=0.1)
```

```python
model.save("/final_shopper_model.h5")
```
